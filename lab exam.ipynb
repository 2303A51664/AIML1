Prediction of Online Shoppers Purchase Intentions.

 1.Name the product most liked and purchased

 2.Identify the special days when most products are purchased

 3.List the top 5 special days of purchase and type of visitors

 4.Identify the months when most purchases happened

 5.What is the max and min revenue generated by the online shoppers

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/online_shoppers_intention.csv')
print(data.head())
```

```python
most_liked_product = data['ProductRelated'].idxmax()  # Row with maximum ProductRelated
```

```python
most_purchases_days = data[data['Revenue'] == True]['SpecialDay'].value_counts()
```

```python
top_5_special_days = most_purchases_days.head(5).index
top_5_visitors = data[data['SpecialDay'].isin(top_5_special_days) & (data['Revenue'] == True)]['VisitorType'].value_counts()
```

```python
most_purchases_months = data[data['Revenue'] == True]['Month'].value_counts()
```

```python
revenue_by_purchase = data[data['Revenue'] == True]['PageValues']
max_revenue = revenue_by_purchase.max()
min_revenue = revenue_by_purchase.min()
```

```python
print("Most liked product row index:", most_liked_product)
print("Most purchased days:", most_purchases_days)
print("Top 5 special days and visitor types:", top_5_special_days, top_5_visitors)
print("Most purchased months:", most_purchases_months)
print("Maximum Revenue:", max_revenue)
print("Minimum Revenue:", min_revenue)
```
